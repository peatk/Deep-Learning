## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
# application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
drop_columns_df = application_df.drop(columns = ['EIN'])
# drop_columns_df.head()
```

```python
# Determine the number of unique values in each column.
drop_columns_df.nunique()
```

```python
# Look at NAME value counts for binning
name_count = drop_columns_df['NAME'].value_counts()
name_count
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
names_to_replace = name_count[name_count < 5].index.to_list()

# Replace in dataframe
for name in names_to_replace:
    drop_columns_df['NAME'] = drop_columns_df['NAME'].replace(name,"Other")

# Check to make sure binning was successful
drop_columns_df['NAME'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_count = drop_columns_df['APPLICATION_TYPE'].value_counts()
# application_count
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = application_count[application_count < 100].index.to_list()

# Replace in dataframe
for app in application_types_to_replace:
    drop_columns_df['APPLICATION_TYPE'] = drop_columns_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
# drop_columns_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning. You may find it helpful to look at CLASSIFICATION value counts >1
classification_count = application_df['CLASSIFICATION'].value_counts()
classification_count_1 = classification_count[classification_count > 1]
# classification_count_1
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = classification_count[classification_count < 1884].index.to_list()

# Replace in dataframe
for cls in classifications_to_replace:
    drop_columns_df['CLASSIFICATION'] = drop_columns_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
# drop_columns_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_dummies = pd.get_dummies(drop_columns_df).astype(int)
# application_dummies
```

```python
# Split our preprocessed data into our features and target arrays
y = application_dummies['IS_SUCCESSFUL']
X = application_dummies.drop(columns = 'IS_SUCCESSFUL')

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Optimization Attempt #1

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
opt1_input_dim = len(X_train_scaled[0])
opt1_hidden_nodes_L1 = 10
opt1_hidden_nodes_L2 = 20 


opt1_nn = tf.keras.models.Sequential()

# First hidden layer
opt1_nn.add(tf.keras.layers.Dense(units = opt1_hidden_nodes_L1, input_dim = opt1_input_dim, activation = 'relu'))

# Second hidden layer
opt1_nn.add(tf.keras.layers.Dense(units = opt1_hidden_nodes_L2, activation = 'relu'))

# Output layer
opt1_nn.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

# Check the structure of the model
opt1_nn.summary()
```

```python
# Compile the model
opt1_nn.compile(loss = 'binary_crossentropy', optimizer= 'adam', metrics=['accuracy'])
```

```python
# Train the model
opt1_fit_model = opt1_nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = opt1_nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Optimization Attempt #1 - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
